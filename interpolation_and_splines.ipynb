#####(a) Complete the following function:

```python
# 1(a)

def lemniscate(d,f):

  theta = np.linspace(0, 2 * np.pi, d)

  r = f * np.sqrt(2 * np.cos(2 * theta))
  x = r * np.cos(theta)
  y = r * np.sin(theta)

  plt.figure(figsize=(8, 8))
  plt.plot(x, y, label="lemniscate")
  plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
  plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
  plt.scatter([-f, f], [0, 0], color='red', label="focuses (F1, F2)")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("Lemniscate of Bernoulli")
  plt.legend()
  plt.grid()
  plt.axis('equal')
  plt.show()
  pass
```

1(c) *The graph was expected to be a  continuous graph, but in the graph above there is a "hole" in the middle.
The reason for that is that cos(theta) can be a negative number which doesn't have a real square value. the solution is to compute r^2 and filter only valid points (=replace negative valuse with zero).*

```python
#code that solves the problem for each parameter
import numpy as np
import matplotlib.pyplot as plt

def lemniscate2(d, f):
    theta = np.linspace(0, 2 * np.pi, d)

    #updated code: compute r^2 and filter only valid points
    r_squared = 2 * f**2 * np.cos(2 * theta)
    r = np.sqrt(np.maximum(r_squared, 0))  #replace negative values with zero


    x = r * np.cos(theta)
    y = r * np.sin(theta)

    plt.figure(figsize=(8, 8))
    plt.plot(x, y, label="Lemniscate")
    plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
    plt.scatter([-f, f], [0, 0], color='red', label="focuses (F1, F2)")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Lemniscate of Bernoulli")
    plt.legend()
    plt.grid()
    plt.axis('equal')
    plt.show()
    pass

#demonstrate that the solution is correct:
lemniscate2(1000, 5)
```

#####(a) Complete the following function:

```python
# 2(a)
def torus(r, R):

  theta = np.linspace(0, 2 * np.pi, 100)
  phi = np.linspace(0, 2 * np.pi, 100)
  theta, phi = np.meshgrid(theta, phi)

  x = (R + r * np.cos(phi)) * np.cos(theta)
  y = (R + r * np.cos(phi)) * np.sin(theta)
  z = r * np.sin(phi)

  fig = plt.figure(figsize=(10, 8))
  ax = fig.add_subplot(111, projection='3d')

  ax.plot_surface(x, y, z, alpha=0.8, edgecolor='k', cmap='Blues')

  ax.set_title("3D torus", fontsize=16)
  ax.set_xlabel("X")
  ax.set_ylabel("Y")
  ax.set_zlabel("Z")
  plt.show()
  pass
```

#####(b) Draw a torus with r=R=1

```python
# 2(b)
torus(1,1)
```
